```python
import pandas as pd
import matplotlib.pyplot as plt


# Example 2: Scatter Plot - Membership Duration vs Revenue

plt.figure(figsize=(8,6))
plt.scatter(gym["Membership_Duration_Months"], gym["Total_Revenue"], color='orange', alpha=0.6)
plt.title("Membership Duration vs Total Revenue")
plt.xlabel("Membership Duration (Months)")
plt.ylabel("Total Revenue")
plt.grid(True)
plt.show()
```

```python
 # 2 Revenue by Membership Type

plt.figure(figsize=(10,6))
gym.groupby('abonoment_type')['gender'].sum().sort_values(ascending=False).plot(kind='bar', color='coral', edgecolor='black')
plt.title('Gender by Membership Type', fontsize=14, weight='bold')
plt.xlabel('Membership Type')
plt.ylabel('Gender')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

```

```python


# 3 Gender Distribution (Pie Chart)
plt.figure(figsize=(6,6))
gender_counts = gym['gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['#ff69b4', '#1e90ff'], startangle=90)
plt.title('Gender Distribution', fontsize=14, weight='bold')
plt.tight_layout()
plt.show()

```

```python
gym.columns
```

```python


# 4 Retention Rate by Subscription Type

plt.figure(figsize=(10,6))
gym.groupby('abonoment_type')['visit_per_week'].mean().sort_values(ascending=False).plot(kind='bar', color='mediumseagreen', edgecolor='black')
plt.title('Retention Rate by Membership Type', fontsize=14, weight='bold')
plt.xlabel('Membership Type')
plt.ylabel('Average Retention Rate (%)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()


```

```python
 

# 5 Average Age by Membership Type (Horizontal Bar)

plt.figure(figsize=(10,6))
gym.groupby('abonoment_type')['Age'].mean().sort_values().plot(kind='barh', color='slateblue', edgecolor='black')
plt.title('Average Age by Membership Type', fontsize=14, weight='bold')
plt.xlabel('Average Age')
plt.ylabel('Membership Type')
plt.tight_layout()
plt.show()
```
