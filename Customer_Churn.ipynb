```python
#Duplicate
Churn = churn.copy()
```

```python
# Data Cleaning
# Drop identified columns(they dont add analytical value)
Churn_clean = Churn.drop(columns=['RowNumber', 'CustomerId','Surname'])

# check for duplicates
print('Duplicates:', Churn_clean.duplicated().sum())

#Ensure no missing values remain
Churn_clean.info()
```

# Data Exploration

```python
# Univariate Analysis (Histograms)

numeric_cols = ['CreditScore','Age','Tenure','Balance','NumOfProducts','EstimatedSalary']

for col in numeric_cols:
    plt.Figure(figsize=(6,3))
    plt.hist(Churn_clean[col], bins=20, color='darkblue', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
```

```python
# 8 Categorial Feature Counts

cat_cols = ['Geography','Gender','HasCrCard','IsActiveMember','Exited' ]
for col in cat_cols:
    plt.Figure(figsize=(5,3))
    counts = Churn_clean[col].value_counts()
    plt.bar(counts.index.astype(str), counts.values, color='lightgreen', edgecolor='black')
    plt.title(f'Count by {col}')
    plt.xlabel(col)
    plt.ylabel('count')
    plt.tight_layout()
    plt.show()
```

```python
# Bivariate Analysis (Churn vs Gender & Geography)

churn_by_genger = Churn_clean.groupby('Gender')['Exited'].mean()
plt.Figure(figsize=(6,8))
plt.bar(churn_by_genger.index, churn_by_genger.values, color=['orange', 'blue'], edgecolor='black')
plt.title('Average Churn Rate by Gender')
plt.ylabel('Churn Rate')
plt.tight_layout()
plt.show()

churn_by_geo = Churn_clean.groupby('Geography')['Exited'].mean()
plt.figure(figsize=(7,5))
plt.bar(churn_by_geo.index,churn_by_geo.values, color=['lightgreen', 'skyblue', 'darkgrey'], edgecolor='black')
plt.title('Average Churn Rate by Geography')
plt.xlabel('Geograghy')
plt.ylabel('Churn Rate')
plt.tight_layout()
plt.show()
```

```python
# Keep only numeric columns
corr = Churn_clean.select_dtypes(include=['number']).corr()

# Plot the heatmap
plt.Figure(figsize=(8,6))
plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)),corr.columns)
plt.title('Correlation Heatmap (Numeric Columns Only)')
plt.tight_layout()
plt.show()

# Show correlation with target variables
corr['Exited'].sort_values(ascending=False)
```
