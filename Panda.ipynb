Execute SQLQuery and Load into Dataframe:
Once the connnection is established, you can use the pd.read_sql to executea SQL query and load the results directly into a
Pandas Dataframe.

Query:The query variable holds the SQL statement you want to execute to retrieve data. You can select specific columns,apply
WHERE clauses, JOIN tables etc. Just like a standard SQL query.

```python
query = "select * from menu"
menu = pd.read_sql_query(query, engine)
menu
```

```python
query = "select * from sales"
sales = pd.read_sql_query(query, engine)
sales
```

```python
query = "select * from members"
members = pd.read_sql_query(query, engine)
members
```

```python
# Step 1 Join sales and members on 'customer_id'
merged_df = pd.merge(sales,members, on='customer_id', how= 'left')
merged_df
```

```python
# Step 2: Join the result with menu on 'product_id'
Dannys = pd.merge(merged_df,menu, on='product_id', how='left')
Dannys
```

```python
# Pivot table
pd.pivot_table(Dannys, values='price', index='customer_id', columns='product_name', aggfunc='max' )
```

```python
pd.pivot_table(Dannys, values='price', index='order_date', columns='product_name', aggfunc='sum' )
```
